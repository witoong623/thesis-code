```python
import cv2
import numpy as np

from PIL import Image
```

```python
def center_crop(img, desired_size, shift_H=1, shift_W=1):
    H, W = img.shape[:2]
    desired_H, desired_W = desired_size
    
    H_cen = int(H // 2 * shift_H)
    W_cen = int(W // 2 * shift_W)
    
    y = H_cen - desired_H // 2
    x = W_cen - desired_W // 2
    
    return np.ascontiguousarray(img[y:y+desired_H, x:x+desired_W])
```

```python
img = cv2.imread('observation_image_0001.jpeg')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
desired_H, desired_W = 384, 768
```

```python
croped_img = center_crop(img, (desired_H, desired_W), shift_H=1.25)
```

```python
original_img = Image.fromarray(img)
crop_img = Image.fromarray(croped_img)
```
